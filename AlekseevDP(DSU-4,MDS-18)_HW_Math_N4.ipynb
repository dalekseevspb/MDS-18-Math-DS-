```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
# Задание №1
# сперва выведем формулу функции f(x) на экран в привычном виде
from sympy import sin,cos
x = sym.Symbol('x')
x**5 + 4*sin(2*x) + cos(3*x+3)
```

```python
# продифференцируем вручную:
# 1-ая производная = 5x^4 + 8cos(2x) - 3sin(3x+3)
# 2-ая производная = 4*5x^3 - 2*8sin(2x) - 3*3cos(3x+3) = 20x^3 - 16sin(2x) - 9cos(3x+3)
```

```python
# проверим себя - вычислим формулу 1-ой производной в Python
diff1 = sym.diff(x**5 + 4*sin(2*x) + cos(3*x+3), x, 1)
diff1
```

```python
# рассчитаем значение 1-ой производной в точке x = 1
diff1_ = diff1.subs(x, 1)
print(diff1_)
print(round(diff1_, 2))
```

```python
# проверим себя - вычислим формулу 2-ой производной в Python
diff2 = sym.diff(x**5 + 4*sin(2*x) + cos(3*x+3), x, 2)
diff2
```

```python
# рассчитаем значение 2-ой производной в точке x = 1
diff2_ = diff2.subs(x, 1)
print(diff2_)
print(round(diff2_, 2))
```

```python
# Задание №2
# Постройте график функции, первой и второй производной:
# f(x) = sin^5(2x+1)
# На отрезке [-5; 5]
```

```python
# сперва выведем формулу функции f(x) на экран в привычном виде
sin(2*x+1)**5
```

```python
# создадим отдельный массив с переменными 'x_', чтобы не пришлось потом переопределять
# (т.к. перем.'x' уже использована при визуализации формул)
x_ = np.arange(-5, 5.1, 0.1)
```

```python
y = (np.sin(2*x_+1))**5
```

```python
plt.plot(x_, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# вычислим формулу 1-ой производной в Python
diff1 = sym.diff((sin(2*x+1))**5, x, 1)
diff1
```

```python
# рассчитаем массив со значениями 1-ой производной
y1 = [diff1.subs(x, i) for i in x_]
y1[:3]
```

```python
plt.plot(x_, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('график 1-ой производной')
plt.grid()
plt.show()
```

```python
# вычислим формулу 2-ой производной в Python
diff2 = sym.diff((sin(2*x+1))**5, x, 2)
diff2
```

```python
# рассчитаем массив со значениями 2-ой производной
y2 = [diff2.subs(x, i) for i in x_]
y2[:3]
```

```python
plt.plot(x_, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('график 2-ой производной')
plt.grid()
plt.show()
```
